#### 2. Creating Data Pipeline and Extracting Text from PDF

This cell specifies the path to the PDF bank statement. It then opens the PDF file using `pdfplumber`, iterates through each page, extracts the text, and stores each line of text along with its corresponding page number in a list called `lines`. This step is crucial for converting the unstructured PDF data into a more manageable format.

```python
import requests # To fetch PDF file from GitHub
from io import BytesIO # To turn PDF file into a file-like object in memory

pdf_url = "https://github.com/meanima/data-science-portfolio/releases/download/v1.0.0/BENJAMIN.YAKUB.KADIRI_7080393332_20260101135910.pdf"

response = requests.get(pdf_url) # Fetching the file
response.raise_for_status() # Raise an error if the HTTP request failed (e.g., 404 or 500)

lines = []

with pdfplumber.open(BytesIO(response.content)) as pdf:
    for page_number, page in enumerate(pdf.pages, start = 1):
        text = page.extract_text()

        if not text:
            continue

        for line in text.split("\n"):
            lines.append({
                "page": page_number,
                "line": line.strip()
            })
```

#### 3. Group Lines into Transactions

This cell processes the extracted `lines` to group them into individual transactions. It uses a regular expression to identify lines that start with a date pattern, assuming each new date signifies the start of a new transaction. This helps in reconstructing multi-line transaction descriptions into single entries.

```python
import re

transactions = []
current = None

date_pattern = re.compile(r"^\d{2}\s[A-Za-z]{3}\s\d{4}")

for item in lines:
    text = item["line"]

    if date_pattern.match(text):
        if current:
            transactions.append(current)

        current = {
            "page": item["page"],
            "raw": text
        }

    else:
        if current:
            current["raw"] += " " + text

if current:
    transactions.append(current)
```

#### 4. Create DataFrame from Raw Transactions

This cell converts the list of raw transactions into a pandas DataFrame. Each row in this DataFrame initially represents a raw transaction string extracted from the PDF. This DataFrame will be the basis for further structured data extraction.

```python
df = pd.DataFrame(transactions)
```

#### 6. Define `split_transaction` Function

This cell defines a helper function, `split_transaction`, which is designed to parse a single raw transaction string and extract structured information such as date, time, description, debit amount, credit amount, and transaction channel. It uses string manipulation and regular expressions to identify and extract these specific components from the raw text.

```python
def split_transaction(row):
    parts = row.split()

    date = None
    time = None
    description = row
    debit = None
    credit = None
    channel = None

    # Must have at least a date
    if len(parts) >= 3:
        date = " ".join(parts[:3])
        description = row.replace(date, "").strip()

    # Time
    if len(parts) >= 4 and ":" in parts[3]:
        time = parts[3]
        description = description.replace(time, "").strip()

    # Amounts are always near the end
    amounts = re.findall(r"[\d,]+\.\d{2}", row)

    if len(amounts) == 1:
        debit = amounts[0]

    elif len(amounts) >= 2:
        debit = amounts[0]
        credit = amounts[1]

    for x in amounts:
        description = description.replace(x, "").strip()

    # channel
    for c in ["POS", "ATM", "USSD", "MOBILE", "WEB"]:
        if c in row.upper():
            channel = c
            break

    return date, time, description, debit, credit, channel
```

#### 7. Apply `split_transaction` to DataFrame

This cell applies the `split_transaction` function to the 'raw' column of the DataFrame. The extracted components (date, time, description, debit, credit, channel) are then assigned to new columns in the DataFrame. This transforms the unstructured raw text into a more structured tabular format.

```python
df[["date", "time", "description", "debit", "credit", "channel"]] = (
    df["raw"].apply(lambda x: pd.Series(split_transaction(x)))
)
```


#### 33. Adjust 'Credit' and 'Balance' Based on Conditions

This cell applies conditional logic to adjust the 'Credit' and 'Balance' columns. It sets 'Credit' to 0.00 if 'Credit' equals 'Debit'. Additionally, for other specific conditions (where 'Credit' is not zero/null and 'Balance' equals 'Debit'), it sets 'Balance' to the 'Credit' value. This is a complex data cleaning step to ensure financial figures are accurately represented.

```python
mask_credit_equals_debit = (data["Credit"] == data["Debit"])
data.loc[mask_credit_equals_debit, "Credit"] = 0.00

# For the 'elif data["Credit"]:' part, we need to ensure the first condition was FALSE for these rows
# And that Credit is not zero or NaN.
# Then, check if Balance equals Debit for these specific rows.

# Create a mask for rows where the first condition was FALSE (Credit != Debit)
mask_not_credit_equals_debit = ~mask_credit_equals_debit

# Condition for the elif block: Credit is not null and not zero
mask_credit_truthy = data["Credit"].notna() & (data["Credit"] != 0)

# Condition for the nested if block: Balance equals Debit
mask_balance_equals_debit = (data["Balance"] == data["Debit"])

# Combine all conditions for the second logical block
final_elif_condition_mask = (
    mask_not_credit_equals_debit &
    mask_credit_truthy &
    mask_balance_equals_debit
)

data.loc[final_elif_condition_mask, "Balance"] = data.loc[final_elif_condition_mask, "Credit"]
```
