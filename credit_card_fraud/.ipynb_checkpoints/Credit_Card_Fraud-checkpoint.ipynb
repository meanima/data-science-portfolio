## 1. Establishing the Data Pipeline

Before we can dive into the analysis, we need to bring the data into our workspace. This step is about more than just opening a file; it's about creating a robust connection that works across different environments.

My intention here is to ensure that whether I'm working in **Google Colab**, on a **Linux desktop**, on a **Windows machine**, or even if this notebook is shared with others, the code automatically finds the correct path. It's a bit like setting up a multi-layered safety net;
- If one location isn't available, it gracefully falls back to the next, ensuring our work isn't interrupted by file path errors.

```python
# Importing the Necessary libraries

# Interaction with the operating system and file paths
import os

# Data manipulation and reading CSV files
import pandas as pd

# Handling specific errors when fetching data from a URL
from urllib.error import HTTPError


# Define the primary source URL - GitHub
url = "https://github.com/meanima/Credit-Card-Fraud/releases/download/v1.0/creditcard.csv"

# Default to GitHub url
path = url

# Potential local file paths for fallback
paths = [
    "/content/drive/MyDrive/Datasets/creditcard.csv", # Google Drive path
    "/home/ben/Desktop/Portfolio/Datasets/CreditCardFraud/creditcard.csv", # Linux path
    "C:/Users/Ben/Desktop/Portfolio/Datasets/CreditCardFraud/creditcard.csv" # Windows path
    ]

try:
    # Try reading directly from URL
    df = pd.read_csv(url)
    print("✅✅ ... Data successfully loaded from GitHub.")

except (HTTPError, Exception):
    # If URL fails, iterate through local paths
    print("❌ Load from URL failed.\nAttempting local paths...\n")

    for p in paths:
        if os.path.exists(p):
            df = pd.read_csv(p)
            print(f"\n\n✅✅ ... Data successfully loaded from Local path: {p}")
            break
    else:
        # If local paths not available, attempt to load from Google Drive from Colab.
        print("❌ ... Load from local paths failed: Not in local path environment.\nAttempting to mount Google Drive...\n")
        try:
            # Mount Google Drive in Colab
            from google.colab import drive # To communicate with Google drive and retrieve our data
            from google.colab._message import MessageError # Handling Message Errors if unable to authenticate Google drive account
            drive.mount("/content/drive")

            for p in paths:
                if os.path.exists(p):
                    df = pd.read_csv(p)
                    print(f"\n\n✅✅ ...  Data successfully loaded from Google Drive.")
                    break
            else:
                # If Mounting from Drive fails
                print("❌ ... Load from Google Drive failed.\n\n\n")

        except MessageError:
            print("Error ❌ ❌: Google Drive Permission denied.\nNo data source found.")
```

## 4.0 Feature Engineering

### 4.1 Creating a Financial Profile using the `'Amount'` Feature

Is a fraudster more likely to make a massive purchase or a series of small ones? To find out, we need to compare the financial signatures of legitimate and fraudulent transactions.

But first, we'll need to separate the data according to the `Class`.

#### 4.1.1 Segregating the Data

To perform a comparative analysis, we first need to split our dataset into two distinct groups based on the transaction type:
- **Normal Transactions:** Identified by the class `0`.
- **Fraudulent Transactions:** Identified by the class `1`.

By isolating these groups, we can calculate specific metrics (like average amounts) for each to identify patterns that distinguish fraud from legitimate activity.

```python
# Filter for normal transactions
normal = class_data.filter(like = "0", axis = 0)

# Filter for fraudulent transactions
fraud = class_data.filter(like = "1", axis = 0)
```

#### 4.1.2 Setting up Profiling

In this step, I'm calculating the average, maximum, and minimum transaction amounts for both groups. This comparison helps us understand if `Amount` is a strong enough indicator on its own to flag suspicious activity, or if fraudsters are clever enough to blend in with the typical spending habits of a normal cardholder.

```python

# Average normal
avg_nrm = normal["Amount"].mean().round(2)
# Average fraud
avg_frd = fraud["Amount"].mean().round(2)

# Max normal
max_nrm = normal["Amount"].max()
#Max fraud
max_frd = fraud["Amount"].max()

# Minimum normal
min_nrm = normal["Amount"].min()
# Minimum fraud
min_frd = fraud["Amount"].min()

# Printing results
print(
    f"Average Normal Transaction = ${avg_nrm}\n",
    f"Average Fraud Transaction = ${avg_frd}\n\n",
    f"Maximum Normal Transaction = ${max_nrm}\n",
    f"Minimum Normal Transaction = ${min_nrm}\n\n",
    f"Maximum Fraud Transaction = ${max_frd}\n",
    f"Minimum Fraud Transaction = ${min_frd}"
)
```

- AVERAGES:
>Normal transactions average `$88.41`, while Fraudulent transactions average `$123.87`.`

This means fraud transactions are **slightly larger** on average. But the difference is not huge, so `Amount` alone cannot perfectly separate fraud from normal transactions.


- MAXIMUMS:
>Normal transactions maxes at `$25,691.16`, while Fraudulent transactions maxes at `$2.125.87`.

Normal transactions can be **much bigger** than fraud ones. This tells us that **big amounts are not always fraud**. Sometimes legitimate purchases are very large. for example, buying a house or car.

- MINIMUMS:
>Both Fraud and Normal transactions have `0.0` as minimum.

Fraudsters also test with **tiny transactions** like `$0 or $1` to see if the card works before making bigger purchases.

What this tells us is that `Amount` is useful, but not enough by itself. We need to combine it with other clues, for example, `Time of Day` or `PCA` features to get better insights.

### 4.2 Creating a Financial Profile using the `'Time'` Feature

#### 4.2.1 Extracting the Rhythm of the Day

While knowing the exact second of a transaction is precise, it's often more useful to look at the broader pattern. By extracting the 'Hour' from our timestamp, we can categorize transactions into the 24 hours of the day.

```python
# Extracting hour from the Time column
class_data["Hour"] = class_data["Time"].dt.hour
```

- This is a key step for behavioral analysis. It allows us to move beyond individual events and start looking for trends: Do we see more fraud during the quiet hours of 3 AM, or does it spike during the high-volume traffic of the business day?

#### 4.2.2 Aggregating Transactions by Hour

In this step, we prepare our data for behavioral analysis by time:
- **Grouping:** We group the records by both 'Hour' and 'Nature' to see the volume of transactions for every hour of the day.
- **Counting:** We calculate the frequency of occurrences in each group.
- **Indexing:** We set 'Nature' as the index so we can easily filter between 'Normal' and 'Fraud' datasets in subsequent steps.

```python
# Grouping the data by 'Hour' and 'Nature'.
hourly_data = class_data.groupby(["Hour", "Nature"]).size().reset_index(name ="Count")

# Setting 'Nature' as the index.
hourly_data = hourly_data.set_index("Nature")
```

#### 4.2.3 Filtering Hourly Patterns

Now that we have the aggregated hourly counts, we split them into two dedicated DataFrames.

```python
# Filtering for fraudulent hourly data
hourly_fraud = hourly_data.filter(like = "Fraud", axis = 0)

# Filtering for normal hourly data
hourly_norm = hourly_data.filter(like = "Normal", axis = 0)
```

- This allows us to analyze the specific behavior and trends of fraudulent transactions versus legitimate ones across the 24-hour cycle.

#### 4.2.4 Identifying Peak Fraud Activity

To understand the intensity of the threat, we calculate the maximum number of fraudulent transactions recorded in any single hour. This helps us identify 'rush hours' for fraudulent activity.

```python
# Identifying peak fraudulent transaction volume hourly
print(
    f"Maximum number of Fraudulent transactions hourly:\n{hourly_fraud['Count'].max()}"
)
```

#### 4.2.4 Identifying Peak Windows of Risk

In this step, we identify the specific times of day when fraudulent activity is most prevalent. By sorting the hourly fraud data in descending order, we can pinpoint the **top 5 hours** with the highest frequency of fraudulent transactions.

```python
# Sorting fraud data by hour and displaying the top 5
hourly_fraud.sort_values(by = "Hour",ascending = False).head(5)
```

#### 4.2.5 Calculating Hourly Financial Impact

To understand the true scale of the threat, we look beyond incident counts.

By grouping the dataset by 'Hour' and 'Nature' to calculate the sum of transaction amounts, we can measure the actual monetary damage caused by fraudulent activity at different times of the day.

```python
# Grouping by hour and nature to sum amounts
hourly_trans = class_data.groupby(["Hour", "Nature"])["Amount"].sum().reset_index(name = "Total").set_index("Nature")

# Filtering for fraud cases
hourly_frd_trans = hourly_trans.filter(like = "Fraud", axis = 0)

# Printing the maximum hourly fraud total
print(
    f"Maximum fraudulent transaction hourly:\n${hourly_frd_trans['Total'].max()}"
)
```

#### 4.2.6 Quantifying the Financial Losses

While frequency tells us how often attacks happen, measuring the `total volume of money stolen` reveals the business impact. We rank these totals to identify the 'Top 5' most expensive hours for the business.

```python
top_fraud = hourly_frd_trans.sort_values(by = "Total", ascending = False)

# Checking the Top 5
top_fraud.head(5)
```

#### 4.2.7 Granular Inspection: Filtering for Specific Hours

To better understand localized activity peaks, we isolate specific hours (such as 11 AM). This deep dive allows us to inspect the specific behavior of transactions within narrow, high-traffic time slices.

## 5.0 The ML Process: Modelling

In the context of our data and the EDA we've performed on it so far, Modeling is the next moral process with which we can build a mathematical algorithm that can automatically distinguish between **'Normal'** and **'Fraudulent'** transactions based on the patterns we've uncovered.

Since our dataset is highly imbalanced:
- Oly 473 fraud cases vs. over 283,000 normal ones;

The model's job is to learn the specific **'Digital signatures'** in the PCA features (V1-V28) and transaction amounts that signal an anomaly.

### 5.0.1 Splitting the Data

Before Modeling can occur on any data, we need to first split it into two:
- X data
- Y data

The model has to learn using our "X data"  and then performs test to try to predict the "Y data" . To achieve that, for each data, we do a train/test on them.

```python
from sklearn.model_selection import train_test_split

# Separate features and target

x = data.drop(["Class", "Nature"], axis = 1) # To remove the answer column from features
y = data["Class"] # To keep answers separately

# Splitting the data
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size = 0.3, # 30% test, 70% train
    random_state = 42, # Same split every time
    stratify = y # Keep fraud ratio consistent
)
```

### 5.1 First Model: Logistic Regression

```python
from sklearn.metrics import roc_auc_score
from sklearn.linear_model import LogisticRegression

# Creating and training the model here to ensure 'model' is defined
lr_model = LogisticRegression(max_iter = 1000) # max_iter = 1000 allows the model to finish learning, no matter how slow the PCA features converges.
lr_model.fit(x_train, y_train)

# Making predictions on unseen data
y_pred = lr_model.predict(x_test)

# Importing the libraries needed to give reports on prediction
from sklearn.metrics import classification_report, confusion_matrix

print(f"Confusion Matrix:\n\n{confusion_matrix(y_test, y_pred)}\n\n\n")
print(f"Classification Report:\n\n{classification_report(y_test, y_pred)}")
```

#### 5.1.1 **Model Evaluation:**

When detecting fraud, a model that is 99% accurate might actually be useless if it misses every single fraud case. Here is how to interpret the results from the previous cell:

#### 1. The Confusion Matrix
This table shows the counts of correct and incorrect predictions:
*   **True Negatives (Top-Left, 84,964):** Legitimate transactions the model correctly identified as Normal.
*   **False Positives (Top-Right, 12):** Normal transactions the model 'falsely' flagged as Fraud (False Alarms).
*   **False Negatives (Bottom-Left, 58):** Fraudulent transactions the model 'missed' (Dangerous error, but one we can work with).
*   **True Positives (Bottom-Right, 84):** Fraudulent transactions the model correctly caught.

#### 2. The Classification Report
Because fraud is rare, we focus on these three metrics for the 'Fraud' class (Class 1):

*   **Precision (0.88):** 'Quality Control.' When the model says a transaction is fraud, it is correct **88%** of the time.
*   **Recall (0.59):** 'The Safety Net.' The model successfully caught **59%** of all actual fraud in the test set. The remaining 11% slipped through. This is manageable.
*   **F1-Score (0.71):** The harmonic mean of Precision and Recall. It gives us a single score to compare models.

Our current Logistic Regression model is very good at avoiding false alarms (high precision), but it is currently missing about 4 out of every 10 fraud attempts (moderate recall).

```python
# Balancing the model with 'class_weight'
balanced_model = LogisticRegression(
    max_iter = 1000,
    class_weight = "balanced",
    random_state = 42
)

# Updating the model's learning - making it balanced
balanced_model.fit(x_train, y_train)

# Making predictions on unseen data
y_balanced_pred = balanced_model.predict(x_test)

print(f"Balanced Confusion Matrix:\n\n{confusion_matrix(y_test, y_balanced_pred)}\n\n\n")
print(f"Balanced Classification Report:\n\n{classification_report(y_test, y_balanced_pred)}")
```

#### 5.1.2 **Understanding the Balanced Model**

In this step, we used the `class_weight='balanced'` parameter to improve our fraud detection.

*   **The Problem:** In the standard model, the algorithm 'sees' so many normal transactions that it becomes biased towards predicting 'Normal' to achieve high accuracy.
*   **The Solution:** By using `balanced`, the model assigns a higher penalty to errors made on the **Fraud** class. It effectively 'multiplies' the importance of the **473** fraud cases so they carry equal weight to the **283,000** normal ones during training.
*   **The Result:** This typically leads to a much higher **Recall** (catching more frauds) at the cost of some **Precision** (more false alarms), which is usually preferable in financial security scenarios.

```python
# Get the predicted probabilities for each class
# [:, 1] isolates the probability of the transaction being 'Fraud' (Class 1)
y_probs = balanced_model.predict_proba(x_test)[:, 1]


import numpy as np

# Define a custom probability threshold to decide if a transaction is fraud
threshold = 0.7

# Classify as 1 (Fraud) only if the predicted probability is greater than or equal to the threshold
y_pred_thresh = (y_probs >= threshold).astype(int)

# Display the confusion matrix to see true/false positives and negatives
print(confusion_matrix(y_test, y_pred_thresh))

# Print the detailed classification report showing precision, recall, and F1-score
print(classification_report(y_test, y_pred_thresh))
```

#### 5.1.3 **Custom Threshold Evaluation**

In this step, we refine the model's decision-making process:

*   **Probability Prediction:** Instead of just getting a 'Normal' or 'Fraud' label, we get the specific probability (score from 0 to 1) for each transaction.
*   **Thresholding (0.7):** We set a higher bar. A transaction is only flagged as 'Fraud' if the model is at least **70% confident**. This helps reduce 'False Alarms'.
*   **Final Metrics:** We use the Confusion Matrix and Classification Report to see how this higher threshold impacts our ability to catch fraud versus the number of legitimate transactions we accidentally flag.

```python
# Raising probability threshold to 90%
threshold = 0.9

# Classify as 1 (Fraud) only if the predicted probability is greater than or equal to the threshold
y_pred_thresh = (y_probs >= threshold).astype(int)

# Display the confusion matrix to see true/false positives and negatives
print(confusion_matrix(y_test, y_pred_thresh))

# Print the detailed classification report showing precision, recall, and F1-score
print(classification_report(y_test, y_pred_thresh))
```

- At threshold 0.5, we catch 89% of fraud but flag ~3,000 transactions.”
- At threshold 0.7, we still catch 89% of fraud but this time, flag ~1500 transactions(about half - significantly lesser).

- At threshold 0.9, we catch 86% of fraud but flag 545 transactions(way less considering the previous number).


Catching 86% of fraud compared to the initial 89%, yet registering a significant 80.8% drop in false negatives (false alarms), We recommend this operating point based on business tolerance. It makes our model stricter, reducing noise, but at the same time doing a pretty decent job of catching 86% fraud transactions.

### 5.3 Gradient Boosting

Gradient Boosting is a powerful machine learning technique that builds models sequentially. It works by combining many 'weak' prediction models (typically decision trees) into a single 'strong' model. In each iteration, it tries to correct the errors of the previous models, fitting new models to the 'residuals' (the differences between the actual and predicted values).

This iterative improvement process allows Gradient Boosting to achieve high accuracy, especially in tasks where complex patterns need to be identified.

```python
from sklearn.ensemble import GradientBoostingClassifier

# Import the necessary functions for ROC curve and AUC score calculations
from sklearn.metrics import roc_curve


gb_model = GradientBoostingClassifier()
gb_model.fit(x_train, y_train)

# Calculate the probability scores for the positive class (Fraud) using the Gradient Boosting model
y_gb_prob = gb_model.predict_proba(x_test)[:, 1]

# Calculate the Area Under the Curve (AUC) for the Gradient Boosting model
gb_auc = roc_auc_score(y_test, y_gb_prob)

print(f"Gradient Boosting ROC-AUC Score: {gb_auc:.4}")
```

### 5.4 Model Performance Comparison: ROC Curve Analysis

The ROC (Receiver Operating Characteristic) curve provides a visual summary of the models' performance across all classification thresholds.

ROC curve plots:

* X-axis → False Positive Rate (FPR)
* Y-axis → True Positive Rate (Recall)

As we move the threshold from 1 to 0, we trace a curve. Each point is a different threshold. So ROC curve shows:
* What happens across ALL possible thresholds.

```python
# Import the matplotlib plotting library
import matplotlib.pyplot as plt


# Calculate the probability scores for the positive class (Fraud) using the Logistic Regression model
y_prob_log = balanced_model.predict_proba(x_test)[:, 1]

# Compute the False Positive Rate and True Positive Rate for the Logistic Regression model
fpr_log, tpr_log, _ = roc_curve(y_test, y_prob_log)

# Calculate the Area Under the Curve (AUC) for the Logistic Regression model
auc_log = roc_auc_score(y_test, y_prob_log)


# Calculate the probability scores for the positive class (Fraud) using the Random Forest model
y_prob_rf = rf_model.predict_proba(x_test)[:, 1]

# Compute the False Positive Rate and True Positive Rate for the Random Forest model
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)

# Calculate the Area Under the Curve (AUC) for the Random Forest model
auc_rf = roc_auc_score(y_test, y_prob_rf)


# Compute the False Positive Rate and True Positive Rate for the Gradient Boosting model
fpr_gb, tpr_gb, _ = roc_curve(y_test, y_gb_prob)


# Initialize a new figure with specific dimensions (6 inches wide by 5 inches high)
plt.figure(figsize=(6,5))

# Plot the ROC curve for Logistic Regression with its corresponding AUC score in the label
plt.plot(fpr_log, tpr_log, label=f'Logistic Regression (AUC={auc_log:.4f})')

# Plot the ROC curve for Random Forest with its corresponding AUC score in the label
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC={auc_rf:.4f})')

# Plot the ROC curve for Gradient Boosting with its corresponding AUC score in the label
plt.plot(fpr_gb, tpr_gb, label=f'Gradient Boosting (AUC={gb_auc:.4f})')

# Plot a diagonal dashed line representing a random classifier (baseline)
plt.plot([0,1], [0,1], [0,1], 'k--')

# Label the X-axis as 'False Positive Rate'
plt.xlabel("False Positive Rate")

# Label the Y-axis as 'True Positive Rate'
plt.ylabel("True Positive Rate")

# Set the title of the ROC Curve chart
plt.title("ROC Curve Comparison")

# Add a legend to the plot to identify each model's curve
plt.legend()

# Display the final plot to the user
plt.show()
```

### 5.5 Precision-Recall Curve

The Precision-Recall (PR) curve is particularly useful for evaluating the performance of binary classifiers on imbalanced datasets, where the positive class (fraud) is rare. It plots Precision (Positive Predictive Value) against Recall (True Positive Rate) at various classification thresholds.

Here's how to interpret the plot:

-   **Precision (Y-axis):** The proportion of positive identifications that were actually correct. High precision means a low false positive rate.
-   **Recall (X-axis):** The proportion of actual positives that were correctly identified. High recall means a low false negative rate.
-   **Goal:** A perfect model would have a PR curve that goes straight up to (0,1) and then across to (1,1), staying in the top-right corner. The closer a curve is to the top-right corner, the better the model performs.

```python
from sklearn.metrics import precision_recall_curve


# Calculate precision and recall values for the Logistic Regression model
precision_log, recall_log, _ = precision_recall_curve(y_test, y_prob_log)

# Calculate precision and recall values for the Random Forest model
precision_rf, recall_rf, _ = precision_recall_curve(y_test, y_prob_rf)

# Calculate precision and recall values for the Gradient Boosting model
precision_gb, recall_gb, _ = precision_recall_curve(y_test, y_gb_prob)


# Set the figure size for the plot
plt.figure(figsize=(6,5))

# Plot the Precision-Recall curve for Logistic Regression
plt.plot(recall_log, precision_log, label="Logistic Regression")

# Plot the Precision-Recall curve for Random Forest
plt.plot(recall_rf, precision_rf, label="Random Forest")

# Plot the Precision-Recall curve for Gradient Boosting
plt.plot(recall_gb, precision_gb, label="Gradient Boosting")

# Label the X-axis as 'Recall' (sensitivity)
plt.xlabel("Recall")

# Label the Y-axis as 'Precision' (positive predictive value)
plt.ylabel("Precision")

# Set the title of the plot
plt.title("Precision-Recall Curve")

# Display the legend to identify the lines
plt.legend()

# Render the plot
plt.show()
```

##### **Interpretating the Precision-Recall Curve**

From the plot, we can observe:

-   **Random Forest:** This model's curve generally stays higher, especially at lower recall values, indicating that it maintains a higher precision for a given recall compared to Logistic Regression and Gradient Boosting. This means Random Forest is better at avoiding false positives while still catching a good proportion of fraudulent transactions.
-   **Logistic Regression:** Its curve is below Random Forest but significantly above Gradient Boosting, showing good performance, but with a potential trade-off. It achieves high recall but might have slightly lower precision than Random Forest at certain operating points.
-   **Gradient Boosting:** The PR curve for Gradient Boosting is very close to the bottom-left corner, indicating poor performance. It has very low precision and recall values, suggesting it's not effectively identifying fraudulent transactions.

**Conclusion:** In the context of fraud detection, where false positives can lead to customer inconvenience and operational costs, a model that maintains high precision (like Random Forest) is often preferred, even if it means a slightly lower recall compared to a model like Logistic Regression. The PR curve clearly highlights this trade-off and helps in selecting the most appropriate model based on specific business objectives.

```python
# Get the feature importances from the trained Random Forest model
importances = [
    (np.abs(lr_model.coef_[0]), "Logistic Regression"),
    (rf_model.feature_importances_, "Random Forest"),
    (gb_model.feature_importances_, "Gradient Boosting")
]

fig, axes, = plt.subplots(1, 3, figsize = (15, 5))

for ax, (i, title) in zip(axes, importances):
    # Get the indices of the top 10 most important features, sorted in ascending order
    indices = np.argsort(i)[-10:]

    # Create a horizontal bar chart of the top 10 feature importances
    ax.barh(range(len(indices)), i[indices])
    
    # Set the y-axis tick labels to the names of the top 10 features
    ax.set_yticks(range(len(indices)), x_train.columns[indices])
    
    # Set the title of the plot
    ax.set_title(f"Top 10 Important Features of {title}")
    ax.set_xlabel("Predictive Power")
    ax.set_ylabel("Features")

plt.tight_layout()
# Display the plot
plt.show()
```

**Key observations from these plots are:**

*   **High Importance:** These are the variables that the models' algorithm relies on most heavily to make its predictions. In a fraud detection context, these features likely capture the most distinct patterns or anomalies associated with fraudulent activity.

*   **PCA Components Dominance:** Given the nature of our dataset, where most features are anonymized PCA components (V1-V28), it is expected that several of these 'V' features will appear as highly important. Specifically, `'V14' and 'V10'` are identified as the most crucial features as they appear across all three models and shows the strongest predictive power in the **Random Forest** model, suggesting they contain strong signals that differentiate fraud from legitimate transactions, with `'V4'` trailing behind. `'V12' and 'V17'` also show significant importance across **Random Forest** and **Gradient Boosting** models, indicating strong signals as well.

*   **Actionable Insights:** Understanding these importances helps in several ways:
    *   **Model Interpretability:** It gives us insight into *why* the model makes certain predictions, which is crucial for trust and debugging.
    *   **Feature Selection:** For future iterations or simpler models, we might consider focusing on these highly important features to reduce dimensionality without losing much predictive power.
    *   **Domain Knowledge Validation:** If certain features that are known to be indicative of fraud in real-world scenarios appear as important here, it validates the model's learning process.

This plot reinforces our earlier exploratory analysis (EDA) where certain 'V' features demonstrated clearer separation between 'Normal' and 'Fraud' distributions, confirming their value for the model.
